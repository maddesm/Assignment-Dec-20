```python
# part 1 : bioinformatics 
## Overlap Graphs (grph)

# input is multiple lists of dns called dna_list
# output is the overlaps of the dna lists
# so just go through each index by index and add to overlaps list
# Read a FASTA file and return a list of sequences 

from Bio import SeqIO

# Parse the FASTA file
def ReadFASTA(filename):
    """Parses a FASTA file using Biopython and returns a list of (label, sequence) tuples."""
    sequences = []
    for record in SeqIO.parse(filename, 'fasta'):
        sequences.append((record.id, str(record.seq)))
    return sequences

dna_list = ReadFASTA('rosalind_grph.txt')

# Find overlaps
overlaps = []
for i in range(len(dna_list)):
    for j in filter(lambda j: j != i, range(len(dna_list))):
        if dna_list[i][1][-3:] == dna_list[j][1][0:3]:
            overlaps.append(dna_list[i][0] + ' ' + dna_list[j][0])
print('\n'.join(overlaps))
```

```python
## tree 
# ( Completing a Tree )- The Tree of Life 

# input is number of nodes = n 
# output :the number of additional edges needed (MINIMUM)
with open('rosalind_tree.txt') as input_data:
    edges = input_data.read().strip().split('\n')
n = int(edges.pop(0))
processed_edges = [] # pairing
for edge in edges:
    processed_edges.append([int(x) for x in edge.split()])

# connected the nodes into a list
individual_nodes = [{i} for i in range(1, n + 1)]

# use edges to put them togather 
# but stop  when to exit ? when both are there ( the question asks the minimum number)
for edge in processed_edges:
    temp_nodes, del_nodes = set(), []
    found = False
# now update the list (individual nodes take i made from connected nodes)
    for nodes in individual_nodes:
        if not found:
            if edge[0] in nodes and edge[1] in nodes:
                found = True
            elif edge[0] in nodes or edge[1] in nodes:
                temp_nodes.update(nodes)
                del_nodes.append(nodes)
                if len(del_nodes) == 2:
                    found = True
    if del_nodes:
        temp_nodes.update(edge)
        # now remove the old nodes and add the merged nodes 
        for nodes in del_nodes:
            individual_nodes.remove(nodes)
        individual_nodes.append(temp_nodes)

# the answer should be the number of elements in the list 
length = int(len(individual_nodes))
print(length-1)
```

```python
## Long
# Genome Assembly as Shortest Superstring long)
# input should be imported as fasta 
from Bio import SeqIO
def ReadFASTA(filename):
    sequences = []
    for record in SeqIO.parse(filename, 'fasta'):
        label = record.id 
        # the sequence should be turend into a string the added to the list with the label and its sequence
        sequence = str(record.seq)
        sequences.append((label, sequence))
    return sequences

# Maximum overlap
def merge_max_overlap(str_list):
    """Merge two strings with the maximum overlap."""
    max_length = -1
    max_indices = None

    for i in range(len(str_list)):
        for j in range(len(str_list)):
            if i != j:  # Avoid self-comparison
                prefix, suffix = str_list[i], str_list[j]
                overlap_len = 0
                for k in range(1, len(prefix)):
                    if prefix[k:] == suffix[:len(prefix) - k]:
                        overlap_len = len(prefix) - k
                        break
                
                if overlap_len > max_length:
                    max_length = overlap_len
                    max_indices = (i, j)

    # Merge strings based on max overlap
    i, j = max_indices
    merged_string = str_list[i] + str_list[j][max_length:]
    return [str_list[k] for k in range(len(str_list)) if k not in max_indices] + [merged_string]

def longest_common_superstring(str_list):
    """Construct the longest common superstring."""
    while len(str_list) > 1:
        str_list = merge_max_overlap(str_list)
    return str_list[0]

# Main execution
if __name__ == '__main__':
    dna_list = ReadFASTA('rosalind_long.txt')  # Read data
    sequences = [seq for _, seq in dna_list]  # Extract only sequences

    # Compute the shortest superstring
    super_string = longest_common_superstring(sequences)

    # Print the result
    print(super_string)

```

```python
#deg (Degree Array)
# input sjould be as a fasta formatting, input the pile as n and m 
# and a list of edges 

def read_graph_from_file(file_path):
    with open(file_path, 'r') as f:
        n, m = map(int, f.readline().strip().split())
        edges = [tuple(map(int, line.strip().split())) for line in f]
    return n, edges

# now using networkX I can calculate the degree of each vertex
def compute_degrees_with_networkx(edges):
    # Create a graph using NetworkX
    G = nx.Graph()
    G.add_edges_from(edges)

    # Compute degrees and return as a 1-indexed list
    degrees = [G.degree[node] for node in range(1, len(G.nodes) + 1)]
    return degrees

def main():
    file_path = 'rosalind_deg.txt'
    n, edges = read_graph_from_file(file_path)
    degrees = compute_degrees_with_networkx(edges)
    print(' '.join(map(str, degrees)))

if __name__ == "__main__":
    main()
```

```python
## Bfs (Breadth-First Search)
data = "rosalind_bfs.txt"
f = open(data, "r")
vertice, edge = map(int, f.readline().strip().split(" "))
graph = {i+1:[] for i in range(vertice)}
for line in f:
    l = list(map(int, line.strip().split(" ")))
    graph[l[0]].append(l[1])
f.close()

# shortest path
def BFS(start_vertice, vertice, graph):
    quene, order = [], [] 
    distance = {i+1:0 for i in range(vertice)} 
    quene.append(start_vertice)
    order.append(start_vertice)
    while quene:
        v = quene.pop(0)
        for n in graph[v]:
            if n not in order:
                distance[n] = distance[v] + 1
                order.append(n)
                quene.append(n)
    for k in distance.keys():
        if k not in order:
            distance[k] = -1
    return order, distance

start_vertice = 1
order, distance = BFS(start_vertice, vertice, graph)
for i in range(int(vertice)):
    print(distance[i+1], end = " ")
```

```python
## dij (Dijkstra's Algorithm)

import numpy as np
import sys

data = "rosalind_dij.txt"
with open(data, "r") as f:
    vertice, edge = map(int, f.readline().strip().split(" "))
    graph = [[sys.maxsize] *vertice for i in range(vertice)]
    for line in f:
        node1, node2, weight = map(int, line.strip().split(" "))
        graph[node1-1][node2-1] = weight

def minDistance(dist, visited):
    mini = sys.maxsize
    mini_index = dist.index(mini)
    for v in range(vertice):
        if dist[v] < mini and visited[v] == False:
            mini = dist[v]
            mini_index = v
    return mini_index

def dijkstra(src, graph, vertice):
    dist = [sys.maxsize] * vertice 
    dist[src] = 0 
    visited = [False] * vertice 

    for c in range(vertice):
        u = minDistance(dist, visited)
        visited[u] = True

        for v in range(vertice):
            if graph[u][v]>0 and visited[v]==False and dist[v] > dist[u]+graph[u][v]:
                dist[v] = dist[u] + graph[u][v]

    for n in range(vertice):
       

        if dist[n] < sys.maxsize:
            print(dist[n], end=" ")
        else:
            print(-1, end=" ")

dijkstra(0, graph, vertice)
```
